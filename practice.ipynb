```python
#whether the character is alphabet, digit or special character
ch= input("Enter a character: ")

if ch.isalpha():
    print("It's an alphabet")
elif ch.isdigit():
    print("It's a digit")
else:
    print("It's a special character")
```

```python
# #reverse a number

num=int(input("Enter a number:"))
r=0
while num>0:
    d=num%10 #12
    r=r*10+d #4
    num=num//10 #123
print(r)
```

```python
#check a number whether palidrome or not
num=int(input("Enter a number:"))
org=num
r=0
while num>0:
    d=num%10 #12
    r=r*10+d #4
    num=num//10 #123
if r==org:
    print("It is a palidrome number")
else:
    print("It is  not a palidrome number")
    
```
